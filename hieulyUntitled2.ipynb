```python
 n = int(input("Nhập số lượng n: "))  # Nhập số lượng n
tong = 0

for i in range(n):
    x = float(input(f"Nhập số thứ {i+1}: "))  # Nhập từng số
    tong += x

trung_binh = tong / n  # Tính trung bình cộng
print("Trung bình cộng của", n, "số là:", trung_binh)
```

```python
n = int(input("Nhập số lượng n: "))
dem_chan = 0

for i in range(n):
    x = int(input(f"Nhập số thứ {i+1}: "))
    if x % 2 == 0:
        dem_chan += 1

print("Số lượng số chẵn trong danh sách là:", dem_chan)
```

```python
n = int(input("Nhập số lượng n: "))
danh_sach = []

for i in range(n):
    x = float(input(f"Nhập số thứ {i+1}: "))
    danh_sach.append(x)

max_gia_tri = danh_sach[0]
min_gia_tri = danh_sach[0]

for so in danh_sach:
    if so > max_gia_tri:
        max_gia_tri = so
    if so < min_gia_tri:
        min_gia_tri = so

print("Giá trị lớn nhất là:", max_gia_tri)
print("Giá trị nhỏ nhất là:", min_gia_tri)
```

```python
print("Nhập 5 số:")
danh_sach = []

for i in range(5):
    x = float(input(f"Số thứ {i+1}: "))
    danh_sach.append(x)

# Sắp xếp tăng dần bằng thuật toán nổi bọt (bubble sort)
for i in range(len(danh_sach)):
    for j in range(0, len(danh_sach) - i - 1):
        if danh_sach[j] > danh_sach[j + 1]:
            danh_sach[j], danh_sach[j + 1] = danh_sach[j + 1], danh_sach[j]

print("Danh sách sau khi sắp xếp tăng dần là:")
print(danh_sach)
```

```python
def giaithua(n):
    if n == 0:  # Điều kiện dừng (0! = 1)
        return 1
    else:
        return n * giaithua(n-1)  # Gọi lại hàm để tính giai thừa

n = int(input("Nhập số n để tính giai thừa: "))
result = giaithua(n)
print(f"Giai thừa của {n} là: {result}")
```
