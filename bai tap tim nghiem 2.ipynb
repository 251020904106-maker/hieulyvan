```python
n = int(input("Enter a positive integer: "))
print(n)

s = 0  # biến số dùng để lưu tổng

i = 2  # bắt đầu từ số chẵn đầu tiên
chuoi = ""
while i < n:
    s += i      # cộng dồn vào tổng
    chuoi += str(i) + " + "
    i += 2      # chuyển sang số chẵn tiếp theo

print(f"The sum of even numbers less than {chuoi} is {s}")
```
