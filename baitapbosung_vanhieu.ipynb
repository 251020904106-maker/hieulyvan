```python
nam = int(input("Nhập một năm bất kỳ: "))
print("\n--- KIỂM TRA NĂM NHUẬN ---")
if (nam % 4 == 0 and nam % 100 != 0):
    print(nam, "là năm nhuận vì chia hết cho 4 và không chia hết cho 100.")
elif nam % 400 == 0:
    print(nam, "là năm nhuận vì chia hết cho 400.")
else:
    print(nam, "KHÔNG phải năm nhuận.")
```

```python
n = int(input("Nhập số nguyên dương n: "))
tong = 0  # Khởi tạo tổng ban đầu = 0
for i in range(1, n + 1):
    tong += i  # giống với tong = tong + i
print("Tổng các số từ 1 đến", n, "là:", tong)
```

```python
print("Nhập 10 số nguyên:")
dem_am = 0
dem_duong = 0
dem_0 = 0
for i in range(1, 11):
    so = int(input(f"Nhập số thứ {i}: "))
    if so < 0:
        dem_am += 1
    elif so > 0:
        dem_duong += 1
    else:
        dem_0 += 1
print("\n--- KẾT QUẢ ---")
print("Số âm      :", dem_am)
print("Số dương   :", dem_duong)
print("Số bằng 0  :", dem_0)
```

```python
n = int(input("Nhập số nguyên n: "))
la_nguyen_to = True
if n <= 1:
    la_nguyen_to = False
else:
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            la_nguyen_to = False
            break
if la_nguyen_to:
    print(n, "là số nguyên tố.")
else:
    print(n, "KHÔNG phải là số nguyên tố.")
```
