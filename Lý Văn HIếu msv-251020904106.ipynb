```python
n = int(input("n = "))
tong = 0
for i in range(1, n + 1):
  if i % 2 == 0:
    tong += i
print("tổng số chẵn từ 1 tới",n ,"có tổng là :", tong)
```

```python

s = input("Nhập vào một chuỗi: ")
words = s.split()
word_count = len(words)
longest_word = max(words, key=len)
print(f"words: {word_count} longest word: {longest_word}")
```

```python
nums = list(map(int, input("nhập các số nguyên (cách nhau bưởi dấu cách hoặc dấu gạch ngang): ").split('-')))
nums_unique = list(set(nums))
if len(nums_unique) < 2:
   print("không tồn tại số lớn thứ hai")
else:
   nums_unique.sort()
   second_largest = nums_unique[-2]
print("danh sách sau khi loại trùng và sắp xếp:", nums_unique)
print("số lớn thứ hai là:", second_largest)
```

```python
ds = list(map(float, input("Ds điểm sinh viên").split()))
ds = [d for d in ds if 0<= d <= 10]
if ds:
    print("Điểm trung bình:", sum (ds)/len(ds))
    print("số sinh viên đạt:", sum(d >= 5 for d in ds))
else:
    print("không có điểm hợp lệ")
```
