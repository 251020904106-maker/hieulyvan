```python
import math

# Nhập dữ liệu
s = input("Enter 3 real number (separated by space): ")
s = s.split()
a = float(s[0])
b = float(s[1])
c = float(s[2])

print(f"a = {a}, b = {b}, c = {c}")

# Xử lý
if a == 0:
    # Phương trình bậc nhất bx + c = 0
    if b == 0:
        if c == 0:
            print("Phuong trinh co vo so nghiem.")
        else:
            print("Phuong trinh vo nghiem.")
    else:
        x = -c / b
        print(f"Phuong trinh bac nhat co nghiem x = {x}")
else:
    # Phương trình bậc hai
    delta = b*b - 4*a*c
    print(f"Delta = {delta}")

    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print(f"Phuong trinh co 2 nghiem phan biet: x1 = {x1}, x2 = {x2}")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Phuong trinh co nghiem kep: x = {x}")
    else:
        print("Phuong trinh vo nghiem trong tap so thuc.")
```
